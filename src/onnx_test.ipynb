```python
import onnxruntime as ort
import numpy as np
```

```python
onnx_path = "evagent.onnx"
session = ort.InferenceSession(onnx_path, providers=ort.get_all_providers())
input_name = session.get_inputs()[0].name
print("model input name:" , input_name)
```

```python
obs = np.linspace(-1,1,16, dtype=np.float32).reshape(1,16)
print(session.run(None, {input_name: obs})[0])
```

```python
obs = obs.reshape(1, -1)  # shape: (1, 16)
```

```python
outputs = session.run(None, {input_name: obs})
```

```python
action = outputs[0]
print("Predicted Action:", action)
```

```python
for _ in range(10):
    # realistic value ranges
    soc_hist = np.random.uniform(30, 80, 5)
    vref_hist = np.random.uniform(3, 6, 5)
    vout = vref_hist[-1] + np.random.uniform(-0.2, 0.2)
    err = vref_hist[-1] - vout
    ierr = np.clip(err * 10, -5, 5)

    obs = np.array([
        np.mean(soc_hist)/100,   # normalize roughly like training
        -np.clip(err, -1, 0),    # if that's what feature 2 is
        *soc_hist,
        *vref_hist,
        vout,
        err,
        ierr,
        max(0, 1 - np.mean(soc_hist)/100)
    ], dtype=np.float32).reshape(1,16)

    print(session.run(None, {input_name: obs})[0])
```
